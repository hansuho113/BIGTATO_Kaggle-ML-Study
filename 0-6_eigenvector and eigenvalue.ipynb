#### Let T : V -> V be a linear operator.

#### A nonzero element x ∈ V is called an eigenvector of T if there exists a scalar λ such that Tx = λx.
#### The scalar λ is called the eigenvalue corresponding to the eigenvector x.
-
#### Similarly, if A is an n*n matrix, a nonzero vector x is called an eigenvector of A, If there exists a scalar λ such that Ax = λx 
#### The scalar λ is called the eigenvalue of A corresponding to the eigenvector x.
-
#### Given a linear transform by A, any points on a eigenvector x will remain on the same direction
#### while its magnitude is multiplied by the corresponding eigenvalue : Ax=λx
